## Exerc√≠cio 1: Conversor Simples

Pe√ßa ao usu√°rio uma temperatura em Celsius e converta para Fahrenheit, usando a f√≥rmula:

**F = C √ó 1.8 + 32**

## Exerc√≠cio 4: Troca de Valores

Declare duas vari√°veis, a = 10 e b = 20. Troque os valores entre elas sem usar uma terceira vari√°vel auxiliar. Imprima os novos valores de a e b.


```python
# Solu√ß√£o
a = 10
b = 20

print(f"Antes: a = {a}, b = {b}")

# Troca usando atribui√ß√£o m√∫ltipla
a, b = b, a

print(f"Depois: a = {a}, b = {b}")
```

## Exerc√≠cio 5: Desconto de Produto

Declare uma vari√°vel preco_original = 150.75. Calcule e armazene o valor do pre√ßo com um desconto de 15%. Imprima o valor final.


```python
# Solu√ß√£o
preco_original = 150.75
desconto_percentual = 15
desconto_valor = preco_original * (desconto_percentual / 100)
preco_final = preco_original - desconto_valor

print(f"Pre√ßo original: R$ {preco_original:.2f}")
print(f"Desconto ({desconto_percentual}%): R$ {desconto_valor:.2f}")
print(f"Pre√ßo final: R$ {preco_final:.2f}")
```

## Exerc√≠cio 7: Operador de M√≥dulo

Solicite um n√∫mero inteiro ao usu√°rio. Use o operador de m√≥dulo (%) para verificar se o n√∫mero √© par ou √≠mpar (Dica: o resto da divis√£o por 2 √© 0 para par, e 1 para √≠mpar). Imprima a conclus√£o.


```python
# Solu√ß√£o
numero = int(input("Digite um n√∫mero inteiro: "))

if numero % 2 == 0:
    print(f"O n√∫mero {numero} √© PAR")
else:
    print(f"O n√∫mero {numero} √© √çMPAR")
```

## Exerc√≠cio 8: Express√£o Complexa

Resolva e imprima o resultado da seguinte express√£o:

**(10 + 5 √ó 2) / (3¬≤ - 4) + 1**


```python
# Solu√ß√£o
resultado = (10 + 5 * 2) / (3**2 - 4) + 1
print(f"Resultado: {resultado}")
```

## Exerc√≠cio 9: Incremento e Decremento

Declare uma vari√°vel contador = 5. Use operadores de atribui√ß√£o composta (ex: +=, -=) para adicionar 3 e depois subtrair 1 do valor de contador. Imprima o valor final.


```python
# Solu√ß√£o
contador = 5
print(f"Valor inicial: {contador}")

contador += 3  # Adiciona 3
print(f"Ap√≥s adicionar 3: {contador}")

contador -= 1  # Subtrai 1
print(f"Valor final: {contador}")
```

## Exerc√≠cio 10: Operadores L√≥gicos

Declare chove = True e frio = False. Use operadores l√≥gicos (and, or, not) para imprimir o resultado das seguintes express√µes:

- chove e frio

```python
# Solu√ß√£o
chove = True
frio = False

print(f"chove = {chove}, frio = {frio}")
print(f"chove e frio: {chove and frio}")
print(f"chove ou frio: {chove or frio}")
print(f"n√£o chove: {not chove}")
```
