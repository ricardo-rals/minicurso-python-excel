## Exerc√≠cio 1: Verificador de Idade

Pe√ßa a idade ao usu√°rio. Use if/else para imprimir se a pessoa √© "Maior de idade" (18 anos ou mais) ou "Menor de idade".


```python
# Solu√ß√£o
idade = int(input("Digite sua idade: "))

if idade >= 18:
    print("Maior de idade")
else:
    print("Menor de idade")
```

## Exerc√≠cio 2: Classificador de N√∫mero

Solicite um n√∫mero. Use if/elif/else para classificar e imprimir se o n√∫mero √© positivo, negativo ou zero.


```python
# Solu√ß√£o
numero = float(input("Digite um n√∫mero: "))

if numero > 0:
    print("O n√∫mero √© positivo")
elif numero < 0:
    print("O n√∫mero √© negativo")
else:
    print("O n√∫mero √© zero")
```

## Exerc√≠cio 3: Valida√ß√£o de Senha (Loop while)

Pe√ßa ao usu√°rio para digitar uma senha. Use um loop while para continuar pedindo a senha at√© que a entrada seja exatamente "Python123".


```python
# Solu√ß√£o
senha = ""

while senha != "Python123":
    senha = input("Digite a senha: ")
    if senha != "Python123":
        print("Senha incorreta! Tente novamente.")

print("Senha correta! Acesso permitido.")
```

## Exerc√≠cio 4: Contagem Regressiva (for)

Use um loop for para imprimir uma contagem regressiva de 10 at√© 1.


```python
# Solu√ß√£o
for i in range(10, 0, -1):
    print(i)
print("Lan√ßamento!")
```

## Exerc√≠cio 5: Soma de √çmpares (for e if)

Use um loop for para iterar de 1 a 20. Dentro do loop, use um if para somar apenas os n√∫meros √≠mpares. Imprima a soma total.


```python
# Solu√ß√£o
soma_impares = 0

for numero in range(1, 21):
    if numero % 2 != 0:  # Se o n√∫mero √© √≠mpar
        soma_impares += numero

print(f"A soma dos n√∫meros √≠mpares de 1 a 20 √©: {soma_impares}")
```

## Exerc√≠cio 6: Loop com break

Pe√ßa n√∫meros ao usu√°rio dentro de um loop while infinito. Use a instru√ß√£o break para sair do loop se o usu√°rio digitar o valor 0.


```python
# Solu√ß√£o
while True:
    numero = int(input("Digite um n√∫mero (0 para sair): "))
    if numero == 0:
        print("Saindo do loop...")
        break
    print(f"Voc√™ digitou: {numero}")
```

## Exerc√≠cio 7: Loop com continue

Use um loop for para iterar de 1 a 10. Use a instru√ß√£o continue para pular a impress√£o do n√∫mero 5. Imprima todos os outros.


```python
# Solu√ß√£o
for i in range(1, 11):
    if i == 5:
        continue  # Pula o n√∫mero 5
    print(i)
```

## Exerc√≠cio 8: Faixa de Valores

Pe√ßa uma nota de 0 a 10. Use if/elif/else aninhados ou encadeados para imprimir: "A" (9-10), "B" (7-8.9), "C" (5-6.9), ou "D" (abaixo de 5).


```python
# Solu√ß√£o
nota = float(input("Digite uma nota de 0 a 10: "))

if nota >= 9:
    print("Conceito: A")
elif nota >= 7:
    print("Conceito: B")
elif nota >= 5:
    print("Conceito: C")
else:
    print("Conceito: D")
```

## Exerc√≠cio 9: Tabuada Simples

Solicite um n√∫mero ao usu√°rio. Use um loop for para imprimir a tabuada desse n√∫mero, de 1 a 10 (ex: 5 √ó 1 = 5, 5 √ó 2 = 10, ...).


```python
# Solu√ß√£o
numero = int(input("Digite um n√∫mero para ver sua tabuada: "))

print(f"Tabuada do {numero}:")
for i in range(1, 11):
    resultado = numero * i
    print(f"{numero} √ó {i} = {resultado}")
```

## Exerc√≠cio 10: Login Simples (Tentativas Limitadas)

Crie um programa que permita no m√°ximo 3 tentativas para o usu√°rio adivinhar um n√∫mero secreto fixo (ex: 42). Use um loop for e break se acertar. Se as tentativas acabarem, imprima uma mensagem de falha.


```python
# Solu√ß√£o
numero_secreto = 42
tentativas_maximas = 3

for tentativa in range(1, tentativas_maximas + 1):
    palpite = int(input(f"Tentativa {tentativa}/{tentativas_maximas} - Digite um n√∫mero: "))
    
    if palpite == numero_secreto:
        print("Parab√©ns! Voc√™ acertou!")
        break
    else:
        print("N√∫mero incorreto!")
else:
    print(f"Suas tentativas acabaram! O n√∫mero secreto era {numero_secreto}.")
```
