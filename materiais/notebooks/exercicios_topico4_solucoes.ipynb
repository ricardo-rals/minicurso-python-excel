## Exerc√≠cio 1: Invers√£o de Nome

Pe√ßa o nome completo do usu√°rio. Use slicing de string para imprimir o nome invertido (de tr√°s para frente).


```python
# Solu√ß√£o
nome = input("Digite seu nome completo: ")
nome_invertido = nome[::-1]
print(f"Nome invertido: {nome_invertido}")
```

## Exerc√≠cio 2: An√°lise de Email

Pe√ßa um endere√ßo de e-mail. Verifique e imprima se o e-mail: a) Cont√©m o caractere @ e b) Termina com .com


```python
# Solu√ß√£o
email = input("Digite um e-mail: ")

tem_arroba = email.find('@') != -1
termina_com = email.endswith('.com')

print(f"Cont√©m @: {tem_arroba}")
print(f"Termina com .com: {termina_com}")
```

## Exerc√≠cio 3: Capitaliza√ß√£o

Pe√ßa uma frase. Imprima a frase convertida para t√≠tulo (primeira letra de cada palavra mai√∫scula, usando .title()).


```python
# Solu√ß√£o
frase = input("Digite uma frase: ")
frase_titulo = frase.title()
print(f"Frase em t√≠tulo: {frase_titulo}")
```

## Exerc√≠cio 4: Contagem de Caracteres Espec√≠ficos

Pe√ßa uma frase. Conte e imprima quantas vezes a letra 'a' (min√∫scula ou mai√∫scula) aparece na frase.


```python
# Solu√ß√£o
frase = input("Digite uma frase: ")
contagem = frase.lower().count('a')
print(f"A letra 'a' aparece {contagem} vezes na frase.")
```

## Exerc√≠cio 5: Remo√ß√£o de Espa√ßos

Pe√ßa uma string com espa√ßos extras no in√≠cio e no fim. Use o m√©todo .strip() para remover os espa√ßos e imprima o resultado.


```python
# Solu√ß√£o
texto = input("Digite um texto com espa√ßos no in√≠cio/fim: ")
texto_limpo = texto.strip()
print(f"Texto original: '{texto}'")
print(f"Texto limpo: '{texto_limpo}'")
```

## Exerc√≠cio 6: Formata√ß√£o com format()

Crie as vari√°veis produto = 'Notebook' e preco = 3500.50. Use o m√©todo .format() para imprimir: "O {0} custa R$ {1:.2f}."


```python
# Solu√ß√£o
produto = 'Notebook'
preco = 3500.50
mensagem = "O {0} custa R$ {1:.2f}.".format(produto, preco)
print(mensagem)
```

## Exerc√≠cio 7: Substitui√ß√£o

Pe√ßa uma frase. Substitua todas as ocorr√™ncias da palavra "ruim" por "bom" e imprima o resultado.


```python
# Solu√ß√£o
frase = input("Digite uma frase: ")
frase_substituida = frase.replace("ruim", "bom")
print(f"Frase original: {frase}")
print(f"Frase modificada: {frase_substituida}")
```

## Exerc√≠cio 8: Divis√£o de Frase

Pe√ßa uma frase separada por v√≠rgulas (ex: "ma√ß√£,banana,uva"). Use o m√©todo .split(',') para criar uma lista de palavras e imprima essa lista.


```python
# Solu√ß√£o
frase = input("Digite palavras separadas por v√≠rgula: ")
lista_palavras = frase.split(',')
print(f"Lista criada: {lista_palavras}")
```

## Exerc√≠cio 9: Jun√ß√£o de Palavras

Crie uma lista de palavras ['Programar', 'em', 'Python', '√©', '√≥timo']. Use o m√©todo .join() para juntar as palavras em uma √∫nica frase, separada por espa√ßos.
  

```python
# Solu√ß√£o
palavras = ['Programar', 'em', 'Python', '√©', '√≥timo']
frase = ' '.join(palavras)
print(frase)
```

## Exerc√≠cio 10: Acesso por √çndice

Pe√ßa uma palavra ao usu√°rio. Imprima o primeiro caractere e o √∫ltimo caractere da palavra (usando √≠ndices [0] e [-1]).
