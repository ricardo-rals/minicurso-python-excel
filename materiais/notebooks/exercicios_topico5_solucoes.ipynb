## Exerc√≠cio 1: Cria√ß√£o e Acesso

Crie uma lista de 5 n√∫meros inteiros. Imprima o elemento do meio da lista (√≠ndice 2).

## Exerc√≠cio 2: Adicionar e Remover

Crie uma lista de compras com 3 itens. Use .append() para adicionar um quarto item e .pop() para remover o √∫ltimo item adicionado. Imprima a lista final.


```python
# Solu√ß√£o
compras = ['Arroz', 'Feij√£o', 'Macarr√£o']
print(f"Lista inicial: {compras}")

compras.append('A√ß√∫car')  # Adiciona um item
print(f"Ap√≥s adicionar: {compras}")

compras.pop()  # Remove o √∫ltimo item
print(f"Lista final: {compras}")
```

## Exerc√≠cio 3: Modifica√ß√£o

Crie uma lista de cores. Altere o valor do primeiro elemento da lista para "Preto" e imprima a lista modificada.

## Exerc√≠cio 4: Itera√ß√£o e Dobro

Use um loop for para percorrer a lista numeros = [2, 4, 6, 8]. Imprima o dobro de cada n√∫mero.


```python
# Solu√ß√£o
numeros = [2, 4, 6, 8]

for numero in numeros:
    dobro = numero * 2
    print(f"O dobro de {numero} √© {dobro}")
```

## Exerc√≠cio 5: Lista Vazia e while

Crie uma lista vazia. Use um loop while para pedir ao usu√°rio 3 nomes e adicione-os √† lista. Imprima a lista ap√≥s o loop.


```python
# Solu√ß√£o
nomes = []
contador = 0

while contador < 3:
    nome = input(f"Digite o {contador + 1}¬∫ nome: ")
    nomes.append(nome)
    contador += 1

print(f"Lista de nomes: {nomes}")
```

## Exerc√≠cio 6: Cria√ß√£o de Sublista (Slicing)

Dada a lista letras = ['a', 'b', 'c', 'd', 'e', 'f'], use slicing para criar uma nova lista apenas com os elementos do √≠ndice 1 ao 4 (exclusive). Imprima a sublista.


```python
# Solu√ß√£o
letras = ['a', 'b', 'c', 'd', 'e', 'f']
sublista = letras[1:4]  # Do √≠ndice 1 ao 3 (4 √© exclusive)
print(f"Lista original: {letras}")
print(f"Sublista [1:4]: {sublista}")
```

## Exerc√≠cio 7: Ordena√ß√£o

Crie uma lista de 5 nomes de cidades. Use o m√©todo .sort() para ordenar a lista em ordem alfab√©tica e imprima o resultado.


```python
# Solu√ß√£o
cidades = ['S√£o Paulo', 'Rio de Janeiro', 'Belo Horizonte', 'Salvador', 'Bras√≠lia']
print(f"Lista original: {cidades}")

cidades.sort()
print(f"Lista ordenada: {cidades}")
```

## Exerc√≠cio 8: Invers√£o de Ordem

Dada uma lista de n√∫meros, use o m√©todo .reverse() para inverter a ordem dos elementos e imprima a lista invertida.


```python
# Solu√ß√£o
numeros = [1, 2, 3, 4, 5]
print(f"Lista original: {numeros}")

numeros.reverse()
print(f"Lista invertida: {numeros}")
```

## Exerc√≠cio 9: Verifica√ß√£o de Item

Crie uma lista de filmes. Pe√ßa ao usu√°rio o nome de um filme. Use o operador in para verificar se o filme est√° na lista e imprima a resposta (True ou False).


```python
# Solu√ß√£o
filmes = ['Matrix', 'Inception', 'Interstellar', 'Blade Runner']
filme_busca = input("Digite o nome de um filme: ")

esta_na_lista = filme_busca in filmes
print(f"O filme '{filme_busca}' est√° na lista? {esta_na_lista}")
```

## Exerc√≠cio 10: Listas Aninhadas

Crie uma lista aninhada para representar uma matriz 2x2: [[1, 2], [3, 4]]. Acesse e imprima o n√∫mero 4 (que est√° na segunda linha, segunda coluna).


```python
# Solu√ß√£o
matriz = [[1, 2], [3, 4]]
print(f"Matriz: {matriz}")

# Acessar elemento da segunda linha (√≠ndice 1) e segunda coluna (√≠ndice 1)
numero = matriz[1][1]
print(f"Elemento na segunda linha, segunda coluna: {numero}")
```
