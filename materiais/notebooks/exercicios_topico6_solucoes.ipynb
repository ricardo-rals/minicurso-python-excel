## Exercício 1: Função de Soma

Crie uma função chamada somar que receba dois parâmetros numéricos e retorne a soma deles. Chame a função e imprima o resultado.


```python
# Solução
def somar(a, b):
    return a + b

resultado = somar(5, 3)
print(f"A soma de 5 e 3 é: {resultado}")
```

## Exercício 2: Função Sem Retorno (Procedimento)

Crie uma função chamada saudacao que receba um nome como parâmetro e simplesmente imprima: "Olá, [Nome]! Bem-vindo(a).".


```python
# Solução
def saudacao(nome):
    print(f"Olá, {nome}! Bem-vindo(a).")

saudacao("Maria")
```

## Exercício 3: Função com Valor Padrão

Crie uma função calcular_potencia(base, expoente=2) onde o expoente tem um valor padrão de 2. Teste a função chamando-a apenas com a base (ex: calcular_potencia(5)) e depois com os dois argumentos (ex: calcular_potencia(2, 3)).


```python
# Solução
def calcular_potencia(base, expoente=2):
    return base ** expoente

# Teste 1: apenas com a base (usa expoente padrão = 2)
resultado1 = calcular_potencia(5)
print(f"5² = {resultado1}")

# Teste 2: com base e expoente
resultado2 = calcular_potencia(2, 3)
print(f"2³ = {resultado2}")
```

## Exercício 4: Conversor de Unidades

Crie uma função km_para_milhas(km) que converta quilômetros para milhas (dica: 1 km ≈ 0.621371 milhas). Retorne o valor em milhas.


```python
# Solução
def km_para_milhas(km):
    return km * 0.621371

quilometros = 100
milhas = km_para_milhas(quilometros)
print(f"{quilometros} km = {milhas:.2f} milhas")
```

## Exercício 5: Função de Validação

Crie uma função eh_par(numero) que retorne True se o número for par e False caso contrário. Use o resultado da função dentro de um if.


```python
# Solução
def eh_par(numero):
    return numero % 2 == 0

numero = 8
if eh_par(numero):
    print(f"{numero} é par")
else:
    print(f"{numero} é ímpar")
```

## Exercício 6: Função com Múltiplos Retornos

Crie uma função calcular_operacoes(x, y) que retorne a soma e a subtração dos dois números como uma tupla. Desempacote os resultados em duas variáveis e os imprima.


```python
# Solução
def calcular_operacoes(x, y):
    soma = x + y
    subtracao = x - y
    return soma, subtracao

soma, subtracao = calcular_operacoes(10, 3)
print(f"Soma: {soma}")
print(f"Subtração: {subtracao}")
```

## Exercício 7: Função com Argumentos Arbitrários (*args)

Crie uma função somar_varios(*numeros) que aceite qualquer número de argumentos e retorne a soma total de todos eles.


```python
# Solução
def somar_varios(*numeros):
    total = 0
    for numero in numeros:
        total += numero
    return total

resultado = somar_varios(1, 2, 3, 4, 5)
print(f"A soma é: {resultado}")
```

## Exercício 8: Função de Fatorial

Crie uma função recursiva ou iterativa (for loop) chamada fatorial(n) que calcule e retorne o fatorial de um número inteiro n. Teste com n=5.


```python
# Solução (versão iterativa)
def fatorial(n):
    resultado = 1
    for i in range(1, n + 1):
        resultado *= i
    return resultado

numero = 5
resultado = fatorial(numero)
print(f"O fatorial de {numero} é: {resultado}")
```

## Exercício 9: Escopo de Variável

Declare uma variável global taxa = 0.10. Crie uma função aplicar_taxa(valor) que use essa variável global para calcular valor * taxa. Imprima o resultado.


```python
# Solução
taxa = 0.10  # Variável global

def aplicar_taxa(valor):
    return valor * taxa

valor = 1000
resultado = aplicar_taxa(valor)
print(f"Valor: R$ {valor:.2f}")
print(f"Taxa aplicada (10%): R$ {resultado:.2f}")
```

## Exercício 10: Função Lambda

Crie uma função lambda para calcular o quadrado de um número. Atribua-a a uma variável chamada quadrado e use-a para calcular e imprimir 8².


```python
# Solução
quadrado = lambda x: x ** 2

resultado = quadrado(8)
print(f"8² = {resultado}")
```
