## Exerc√≠cio 4: Filtro Simples

Crie um DataFrame com as colunas 'Pa√≠s' e 'Popula√ß√£o'. Filtre e imprima as linhas onde a 'Popula√ß√£o' √© maior que 1000 (simulado).


````python
# Solu√ß√£o
dados = {
    'Pa√≠s': ['Brasil', 'Argentina', 'Chile', 'Uruguai'],
    'Popula√ß√£o': [215000000, 45000000, 19000000, 3500000]
}

df = pd.DataFrame(dados)
filtrado = df[df['Popula√ß√£o'] > 1000]
print(filtrado)
````

## Exerc√≠cio 5: C√°lculo Estat√≠stico

Crie um DataFrame com uma coluna 'Vendas' (com 5 valores num√©ricos). Calcule e imprima a soma total e a m√©dia das vendas usando os m√©todos do Pandas (.sum(), .mean()).


````python
# Solu√ß√£o
dados = {
    'Vendas': [1000, 1500, 2000, 1800, 2200]
}

df = pd.DataFrame(dados)
soma_total = df['Vendas'].sum()
media = df['Vendas'].mean()

print(f"Soma total: {soma_total}")
print(f"M√©dia: {media:.2f}")
````

## Exerc√≠cio 6: Adi√ß√£o de Coluna

Crie um DataFrame com a coluna 'Sal√°rio'. Adicione uma nova coluna 'B√¥nus' calculada como 20% do 'Sal√°rio'. Imprima o DataFrame com a nova coluna.


````python
# Solu√ß√£o
dados = {
    'Sal√°rio': [5000, 6000, 7000, 5500, 8000]
}

df = pd.DataFrame(dados)
df['B√¥nus'] = df['Sal√°rio'] * 0.20
print(df)
````

## Exerc√≠cio 7: Ordena√ß√£o de Dados

Crie um DataFrame com as colunas 'Nome' e 'Pontua√ß√£o'. Ordene o DataFrame com base na coluna 'Pontua√ß√£o' de forma decrescente (ascending=False) e imprima o resultado.


````python
# Solu√ß√£o
dados = {
    'Nome': ['Ana', 'Bruno', 'Carlos', 'Diana'],
    'Pontua√ß√£o': [85, 92, 78, 95]
}

df = pd.DataFrame(dados)
df_ordenado = df.sort_values('Pontua√ß√£o', ascending=False)
print(df_ordenado)
````

## Exerc√≠cio 8: Agrupamento Simples

Crie um DataFrame com as colunas 'Filial' (A, B, A, B) e 'Receita' (valores num√©ricos). Agrupe os dados por 'Filial' e calcule a soma da Receita para cada filial (.groupby().sum()).


````python
# Solu√ß√£o
dados = {
    'Filial': ['A', 'B', 'A', 'B'],
    'Receita': [10000, 15000, 12000, 18000]
}

df = pd.DataFrame(dados)
agrupado = df.groupby('Filial')['Receita'].sum()
print(agrupado)
````

## Exerc√≠cio 9: Remo√ß√£o de Colunas

Crie um DataFrame. Remova a coluna 'Idade' e imprima o DataFrame restante.


````python
# Solu√ß√£o
dados = {
    'Nome': ['Ana', 'Bruno', 'Carlos'],
    'Idade': [25, 30, 28],
    'Cidade': ['SP', 'RJ', 'BH']
}

df = pd.DataFrame(dados)
print("DataFrame original:")
print(df)

df_sem_idade = df.drop('Idade', axis=1)
print("\nDataFrame sem a coluna 'Idade':")
print(df_sem_idade)
````

## Exerc√≠cio 10: Exporta√ß√£o para Excel
