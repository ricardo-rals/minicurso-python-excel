## Exerc칤cio 1: Gr치fico de Linha B치sico

Crie um gr치fico de linha simples com os dados de X: [1, 2, 3, 4] e Y: [10, 5, 20, 15]. Adicione um t칤tulo.

Crie um gr치fico de barras com categorias: ['A', 'B', 'C'] e valores: [7, 3, 5]. Defina r칩tulos para o eixo X e Y.

## Exerc칤cio 4: Gr치fico de Dispers칚o (Scatter Plot)

Gere um gr치fico de dispers칚o com 10 pontos de dados aleat칩rios para X e Y. Adicione um t칤tulo "Rela칞칚o entre X e Y".


````python
# Solu칞칚o
import random

x = [random.randint(1, 100) for _ in range(10)]
y = [random.randint(1, 100) for _ in range(10)]

plt.scatter(x, y)
plt.title('Rela칞칚o entre X e Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
````

## Exerc칤cio 5: Gr치fico de Pizza B치sico

Crie um gr치fico de pizza para mostrar a distribui칞칚o de 'Despesas' entre [25, 40, 35] para as categorias ['Aluguel', 'Comida', 'Outros'].


````python
# Solu칞칚o
despesas = [25, 40, 35]
categorias = ['Aluguel', 'Comida', 'Outros']

plt.pie(despesas, labels=categorias, autopct='%1.1f%%')
plt.title('Distribui칞칚o de Despesas')
plt.show()
````

## Exerc칤cio 6: Subplots (1x2)

Crie uma figura com dois subgr치ficos lado a lado (1 linha, 2 colunas). Desenhe um gr치fico de linha no primeiro e um gr치fico de barras no segundo.


````python
# Solu칞칚o
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Gr치fico de linha no primeiro subplot
x1 = [1, 2, 3, 4]
y1 = [10, 5, 20, 15]
ax1.plot(x1, y1)
ax1.set_title('Gr치fico de Linha')

# Gr치fico de barras no segundo subplot
categorias = ['A', 'B', 'C']
valores = [7, 3, 5]
ax2.bar(categorias, valores)
ax2.set_title('Gr치fico de Barras')

plt.tight_layout()
plt.show()
````

## Exerc칤cio 7: Grade (Grid)

Crie um gr치fico de linha qualquer e adicione uma grade ao plano de fundo para facilitar a leitura dos valores (plt.grid(True)).

## Exerc칤cio 8: Gr치fico de Barras Horizontais

Use os mesmos dados do Exerc칤cio 2, mas gere um gr치fico de barras horizontais (plt.barh()).

## Exerc칤cio 9: Limites dos Eixos

## Exerc칤cio 10: Legenda
